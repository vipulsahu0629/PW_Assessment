## **Question 1:** What are semantic HTML elements? Why is using them important for web development?

**Answer:**

**Semantic HTML elements** are tags whose names clearly describe the *meaning* or *purpose* of the content they contain — as opposed to generic containers like `<div>` or `<span>` which carry no inherent meaning.

Examples of semantic elements:

| Element | Meaning |
|---------|---------|
| `<header>` | Introductory content or navigation at the top |
| `<nav>` | A block of navigation links |
| `<main>` | The primary/unique content of the page |
| `<article>` | A self-contained piece of content (e.g. a blog post) |
| `<section>` | A thematic grouping of content |
| `<aside>` | Content tangentially related to the main content (sidebar) |
| `<footer>` | Footer for the page or a section |
| `<figure>` / `<figcaption>` | An image or diagram with its caption |

**Why they matter:**
- **Accessibility** — Screen readers use semantic tags to let visually impaired users jump between page regions (e.g. skip straight to `<main>`).
- **SEO** — Search engine crawlers weight content inside meaningful tags (like `<article>` or `<h1>`) more heavily when ranking pages.
- **Readability & maintainability** — A developer reading `<nav>` instantly knows what it is; a `<div id="nav123">` requires guesswork.
- **Browser defaults** — Browsers apply sensible default styles and behaviours to semantic elements, reducing the need for extra CSS.


````python
%%html
<!-- Semantic page skeleton -->
<header><h1>My Blog</h1></header>
<nav><a href='#'>Home</a> | <a href='#'>About</a></nav>
<main>
  <article>
    <h2>My First Post</h2>
    <p>This is the article content.</p>
  </article>
  <aside><p>Related links go here.</p></aside>
</main>
<footer><p>&copy; 2024 My Blog</p></footer>
````

## **Question 3:** How can you make an HTML form more accessible to users with disabilities?

**Answer:**

Accessibility in forms ensures that people using screen readers, keyboard-only navigation, or other assistive technologies can fill them out without frustration. Key techniques:

1. **Use `<label>` with `for` / `id` pairing** — links every label to its control so screen readers announce the label when the field is focused.
2. **Add `placeholder` text** — gives a hint inside the field, but *never* use placeholder as a substitute for a label (it disappears when the user starts typing).
3. **Mark required fields** with `required` attribute and indicate them visually (e.g. asterisk *) and textually.
4. **Use `type` attributes correctly** (`type="email"`, `type="tel"`, `type="date"`) — mobile browsers show the right keyboard; validators catch wrong input early.
5. **Group related fields with `<fieldset>` and `<legend>`** — screen readers read the legend before each field inside the group, giving context (e.g. "Billing Address — Street").
6. **Use `aria-describedby`** to link an input to a hint or error message paragraph.
7. **Ensure keyboard navigability** — all fields and buttons must be reachable and operable with Tab / Enter / Space alone.
8. **Provide clear error messages** — describe *what went wrong* and *how to fix it*, not just "invalid input".


````python
%%html
<style>
  form{font-family:Arial,sans-serif;max-width:360px;padding:16px;background:#f4f4f4;border-radius:8px;}
  label{display:block;margin-top:10px;font-weight:bold;}
  input,select{width:100%;padding:6px;margin-top:4px;box-sizing:border-box;}
  fieldset{border:1px solid #ccc;border-radius:6px;padding:10px;margin-top:10px;}
  button{margin-top:14px;padding:8px 18px;background:#2d6a4f;color:#fff;border:none;border-radius:4px;cursor:pointer;}
</style>
<form>
  <label for='name'>Full Name <span aria-hidden='true'>*</span></label>
  <input type='text' id='name' name='name' required placeholder='e.g. Riya Sharma' aria-required='true'>

  <label for='email'>Email Address <span aria-hidden='true'>*</span></label>
  <input type='email' id='email' name='email' required placeholder='e.g. riya@example.com'>

  <fieldset>
    <legend>Preferred Contact Method</legend>
    <label><input type='radio' name='contact' value='email'> Email</label>
    <label><input type='radio' name='contact' value='phone'> Phone</label>
  </fieldset>

  <button type='submit'>Submit</button>
</form>
````

## **Question 4:** Identify and correct the errors in the following CSS code:

```css
p {
  font-size: 16;
  color: #333
  margin-top 10px;
}
```

**Answer:**

There are **three errors**:

| Line | Error | Fix |
|------|-------|-----|
| `font-size: 16;` | Missing unit — CSS requires a unit for length values. | `font-size: 16px;` |
| `color: #333` | Missing semicolon at the end of the declaration. | `color: #333;` |
| `margin-top 10px;` | Missing colon `:` between property and value. | `margin-top: 10px;` |

Corrected CSS:
```css
p {
  font-size: 16px;
  color: #333;
  margin-top: 10px;
}
```


## **Question 5:** Write CSS rules to style all `<h2>` elements inside a `<section>` with a blue color and center alignment.

**Answer:**

Use a **descendant selector** (`section h2`) which targets any `<h2>` that is a descendant of a `<section>`:


````python
%%html
<style>
section h2 {
  color: blue;
  text-align: center;
}
</style>
<section>
  <h2>This h2 is inside a section — styled!</h2>
  <p>Regular paragraph — unaffected.</p>
</section>
<h2>This h2 is outside a section — NOT styled.</h2>
````

## **Question 6:** Explain the CSS box model and its components.

**Answer:**

Every HTML element is treated as a rectangular **box** made up of four nested layers (from inside out):

1. **Content** — the actual text, image, or child elements. Sized by `width` and `height`.
2. **Padding** — transparent space *inside* the border, between the content and the border. Set with `padding` (or `padding-top`, etc.). Padding takes the element's background color.
3. **Border** — a line drawn around the padding + content area. Set with `border` (style, width, color).
4. **Margin** — transparent space *outside* the border, separating this element from its neighbors. Set with `margin`. Margins are always transparent.
```
┌─────────────────────── Margin ──────────────────────────┐
│  ┌───────────────────── Border ───────────────────────┐ │
│  │  ┌──────────────── Padding ─────────────────────┐  │ │
│  │  │                                              │  │ │
│  │  │               Content                        │  │ │
│  │  │                                              │  │ │
│  │  └──────────────────────────────────────────────┘  │ │
│  └────────────────────────────────────────────────────┘ │
└─────────────────────────────────────────────────────────┘
```


## **Question 7:** How do the `relative`, `absolute`, and `fixed` positioning properties differ in CSS?

**Answer:**

| Property | Reference point | Scrolls with page? | Typical use |
|----------|----------------|--------------------|-------------|
| `position: relative` | The element's *own normal position* | Yes | Nudging an element slightly; creating a positioning context for children |
| `position: absolute` | Nearest *positioned* ancestor(one with `position` ≠ `static`) — or the `<html>` root if none exists | Yes (moves with content) | Tooltips, badges, overlaid labels inside a container |
| `position: fixed` | The **viewport** (browser window) | **No** — stays on screen when scrolling | Sticky headers, floating action buttons, cookie banners |

**`position: sticky`** (bonus): behaves like `relative` until the element reaches a scroll threshold, then acts like `fixed` — great for sticky table headers or nav bars.


````python
%%html
<style>
.parent { position: relative; background: #d8f3dc; padding: 20px; height: 80px; border: 2px solid #2d6a4f; border-radius:6px; margin-bottom:20px; }
.abs-child { position: absolute; top: 8px; right: 8px; background: #f4a261; padding: 4px 8px; border-radius: 4px; font-size:12px; }
.rel-box { position: relative; left: 10px; background: #a8dadc; padding: 10px; border-radius:6px; display:inline-block; }
.fixed-note { position: fixed; top: 10px; right: 10px; background: #e63946; color:#fff; padding: 6px 10px; border-radius:4px; font-size:12px; }
</style>
<div class='parent'>
  <p>Parent (position: relative)</p>
  <span class='abs-child'>Absolute child</span>
</div>
<div class='rel-box'>Relative box (shifted 10px right)</div>
<div class='fixed-note'>Fixed (top-right corner)</div>
````

## **Question 8:** Write a CSS rule to set a background image for a `<div>` with the class `.banner`, ensuring the image covers the entire area without repeating.

**Answer:**

```css
.banner {
  background-image: url('your-image.jpg');
  background-size: cover;       /* scales image to cover the full area */
  background-repeat: no-repeat; /* prevents tiling */
  background-position: center;  /* keeps the focal point centred */
  height: 300px;                /* give the div a visible height */
}
```

Key properties explained:
- `background-size: cover` — scales the image until it completely covers the element's area any overflow is clipped.
- `background-repeat: no-repeat` — stops the image from tiling if it's smaller than the element.
- `background-position: center` — centres the image so the most important part stays visible even after cropping.


````python
%%html
<style>
.banner {
  background-image: url('https://picsum.photos/800/300');
  background-size: cover;
  background-repeat: no-repeat;
  background-position: center;
  height: 400px;
  border-radius: 8px;
}
</style>
<div class='banner'>
  <h2 style='color:#fff;text-shadow:1px 1px 3px #000;padding:20px;'>Welcome to My Site</h2>
</div>
````
