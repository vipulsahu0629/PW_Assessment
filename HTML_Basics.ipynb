## **Question 1:** Explain the basic structure of an HTML page. Why is it important to follow this structure?

**Answer:**

Every HTML page is built from a few core parts:

- `<!DOCTYPE html>` — <!DOCTYPE html> tells the browser this is an HTML5 document.
- `<html>` — the root element that wraps all the content on the page.
- `<head>` — holds information about the page that isn't shown directly to the visitor, such as the `<title>`, character encoding (`<meta charset>`), and links to CSS files.
- `<body>` — It contains everything the user actually sees: headings, text, images, links, forms, etc.

Following this structure matters because:
- Browsers depend on it to parse and render the page correctly and consistently across devices.
- Search engines and screen readers rely on a proper `<head>`/`<body>` split to index and interpret content correctly, which affects SEO and accessibility.
- It keeps code organized and predictable, making it easier for other developers (or future you) to read, debug, and extend the page.

````python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0" />
  <link rel="stylesheet" href="style.css" />
  <title>Browser</title>
</head>
<body>
  <h1>this is <em>HTML</em> boiler-plate code</h1>
</body>
</html>
````

## **Question 3:** Look at the code below. Can you find and fix the mistake?

```html

```

## **Question 4:** You're creating a form that asks the user to enter their name, select their country from a list, and click a button to submit. Write the HTML code for this.

**Answer:**

````python
%%html
<form>
  <label for="name">Name:</label>
  <input type="text" id="name" name="name">

  <label for="country">Country:</label>
  <select id="country" name="country">
    <option value="india">India</option>
    <option value="usa">USA</option>
    <option value="uk">UK</option>
  </select>

  <button type="submit">Submit</button>
</form>
````

## **Question 7:** Here's a form snippet. Can you spot the error and rewrite the correct version?

```html
<form>
  <input name="email">
  <select>
    <option>India</option>
  </select>
  <button>Send</buton>
</form>
```

**Answer:**

There are a few issues:
- The closing tag `</buton>` is misspelled — it should be `</button>`.
- The `<select>` element has no `name` attribute, so its chosen value would never actually be submitted with the form.
- The `<input>` has no `type` attribute — adding `type="email"`.

Corrected version:

````python

%%html
<form>
  <input type="email" name="email" placeholder="Enter your email">
  <select name="country">
    <option>India</option>
  </select>
  <button type="submit">Send</button>
</form>
````
