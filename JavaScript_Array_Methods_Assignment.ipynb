## **Question 1:** Explain what the forEach method does in JavaScript arrays and write code that uses forEach to display each student's name from an array of student objects to the console.
// Array of student objects

```
const students = [
{ name: "Rahul", age: 20 },
{ name: "Priya", age: 19 },
{ name: "Amit", age: 21 },
{ name: "Sneha", age: 20 }
];
```

**Answer:**

`forEach` is an array method that **iterates over every element** and executes a provided callback function for each one. Unlike `map`, it does **not** return a new array — it is used purely for side effects (e.g. logging, updating the DOM).

```
array.forEach(callback(element, index, array))
```

- It always visits every element you cannot `break` out of it early (use `for...of` if you need that).
- It is cleaner and more readable than a traditional `for` loop for simple iteration.


````python
%%javascript
const students = [
  { name: "Rahul", age: 20 },
  { name: "Priya", age: 19 },
  { name: "Amit", age: 21 },
  { name: "Sneha", age: 20 }
];

// forEach with arrow function — print each student's name
console.log("=== Student Names ===");
students.forEach((student, index) => {
  console.log(`${index + 1}. ${student.name}`);
});

// Bonus: forEach with both name and age
console.log("\n=== Name & Age ===");
students.forEach(student => {
  console.log(`Name: ${student.name}, Age: ${student.age}`);
});
````

## **Question 2:** You have an array of product prices in rupees. Write code using the map method to create a new array that contains the same prices but with 18% GST added to each price.
`const originalPrices = [100, 250, 500, 750, 1000];`

**Answer:**

`map` **transforms** every element and returns a **brand-new array** of the same length. The original array is never modified.

- Formula: `price with GST = price × 1.18`


````python
%%javascript
const originalPrices = [100, 250, 500, 750, 1000];

// map returns a new array — original is untouched
const pricesWithGST = originalPrices.map(price => price * 1.18);

console.log("Original prices (₹):", originalPrices);
console.log("Prices with 18% GST:", pricesWithGST);

// Pretty display
console.log("\n=== Price Breakdown ===");
originalPrices.forEach((price, i) => {
  const gst    = (price * 0.18).toFixed(2);
  const total  = (price * 1.18).toFixed(2);
  console.log(`₹${price} + GST ₹${gst} = ₹${total}`);
});
````

## **Question 3:** A school wants to find all students who scored more than 80 marks in their exam. The following code has an error that prevents it from working correctly. Fix the code and explain what was wrong.
```
const students = [
{ name: "Raj", marks: 85 },
{ name: "Sita", marks: 92 },
{ name: "Vikram", marks: 78 },
{ name: "Meera", marks: 88 }
];
const topStudents = students.filter(student => {
student.marks > 80;
});
```
##**Answer:**

**Buggy code:**
```js
const topStudents = students.filter(student => {
  student.marks > 80;   // ❌ BUG: missing return keyword
});
```

**What was wrong:**  
The callback uses **curly braces `{}`** (block body syntax), which means the `return` keyword is required. Without it, the function implicitly returns `undefined`, which is falsy — so `filter` keeps nothing and `topStudents` ends up as an empty array `[]`.

**Two valid fixes:**

| Fix | Syntax |
|-----|--------|
| Add `return` | `student => { return student.marks > 80; }` |
| Remove braces (implicit return) | `student => student.marks > 80` |


````python
%%javascript
const students = [
  { name: "Raj",    marks: 85 },
  { name: "Sita",   marks: 92 },
  { name: "Vikram", marks: 78 },
  { name: "Meera",  marks: 88 }
];

// ❌ BUGGY version — returns empty array
const buggyResult = students.filter(student => {
  student.marks > 80;   // no return!
});
console.log("Buggy result:", buggyResult);  // []

// ✅ FIX 1 — explicit return
const fix1 = students.filter(student => {
  return student.marks > 80;
});
console.log("\nFix 1 (explicit return):", fix1.map(s => s.name)); // Raj, Sita, Meera

// ✅ FIX 2 — concise arrow (implicit return, no braces)
const fix2 = students.filter(student => student.marks > 80);
console.log("Fix 2 (concise arrow):  ", fix2.map(s => s.name)); // same result
````

## **Question 4:** Write code using the find method to locate a book with the title "Harry Potter" from an array of book objects. If the book is found, display its author name, otherwise display "Book not found".
// Array of book objects
```
const books = [
{ title: "Gitanjali", author: "Rabindranath Tagore" },
{ title: "Harry Potter", author: "J.K. Rowling" },
{ title: "The God of Small Things", author: "Arundhati Roy" },
{ title: "Midnight's Children", author: "Salman Rushdie" }
];
```

**Answer:**

`find` returns the **first element** that satisfies the condition, or `undefined` if none match.  
Unlike `filter`, it stops as soon as it finds a match — making it efficient for single lookups.


````python
%%javascript
const books = [
  { title: "Gitanjali",               author: "Rabindranath Tagore" },
  { title: "Harry Potter",            author: "J.K. Rowling" },
  { title: "The God of Small Things", author: "Arundhati Roy" },
  { title: "Midnight's Children",     author: "Salman Rushdie" }
];

// find returns the matching object, or undefined
const found = books.find(book => book.title === "Harry Potter");

// Conditional display using ternary
console.log(found ? `Author: ${found.author}` : "Book not found");

// Test with a title that doesn't exist
const missing = books.find(book => book.title === "Wings of Fire");
console.log(missing ? `Author: ${missing.author}` : "Book not found");

// Bonus: find with destructuring
const { title, author } = books.find(b => b.title === "Gitanjali") || {};
console.log(`\nTitle: ${title}, Author: ${author}`);
````

## **Question 5:** Explain how the sort method works in JavaScript and write code to sort an array of employee objects by their salary in descending order (highest salary first).
// Array of employee objects
```
const employees = [
{ name: "Rajesh", salary: 45000 },
{ name: "Priya", salary: 55000 },
{ name: "Amit", salary: 40000 },
{ name: "Kavya", salary: 60000 }
];
```

**Answer:**

By default, `sort()` converts elements to **strings** and sorts lexicographically — which gives wrong results for numbers (`[10, 9, 100]` → `[10, 100, 9]`). For numeric/custom sorting, pass a **comparator function**:

```
array.sort((a, b) => a - b)   // ascending
array.sort((a, b) => b - a)   // descending
```

The comparator returns:
- **negative** → `a` comes before `b`
- **positive** → `b` comes before `a`
- **0** → order unchanged

> ⚠️ `sort` **mutates** the original array. Use `[...array].sort(...)` to sort a copy.


````python
%%javascript
const employees = [
  { name: "Rajesh", salary: 45000 },
  { name: "Priya",  salary: 55000 },
  { name: "Amit",   salary: 40000 },
  { name: "Kavya",  salary: 60000 }
];

// Sort descending by salary (b - a)
const sorted = [...employees].sort((a, b) => b.salary - a.salary);

console.log("=== Employees by Salary (High → Low) ===");
sorted.forEach((emp, i) => {
  console.log(`${i + 1}. ${emp.name.padEnd(8)} ₹${emp.salary.toLocaleString()}`);
});

// Also show ascending for comparison
const ascending = [...employees].sort((a, b) => a.salary - b.salary);
console.log("\n=== Ascending order ===");
ascending.forEach(e => console.log(`${e.name}: ₹${e.salary}`));
````

## **Question 6:** Write code using the reduce method to calculate the total value of all items in a shopping cart. Each item has a price and quantity. Also find the most expensive single item in the cart.
// Shopping cart with items
```
const cart = [
{ name: "Laptop", price: 50000, quantity: 1 },
{ name: "Mouse", price: 500, quantity: 2 },
{ name: "Keyboard", price: 1500, quantity: 1 },
{ name: "Monitor", price: 15000, quantity: 1 }
];
```
**Answer:**

`reduce` **accumulates** all elements into a single value by applying a callback repeatedly:

```
array.reduce((accumulator, currentValue) => newAccumulator, initialValue)
```

- `accumulator` — the running result carried from one iteration to the next
- `currentValue` — the current element being processed
- `initialValue` — the starting value of the accumulator


````python
%%javascript
const cart = [
  { name: "Laptop",   price: 50000, quantity: 1 },
  { name: "Mouse",    price: 500,   quantity: 2 },
  { name: "Keyboard", price: 1500,  quantity: 1 },
  { name: "Monitor",  price: 15000, quantity: 1 }
];

// Total value = price × quantity for each item, summed
const totalValue = cart.reduce((acc, item) => acc + (item.price * item.quantity), 0);
console.log("Total cart value: ₹" + totalValue.toLocaleString()); // ₹68,000

// Most expensive single item (by unit price)
const mostExpensive = cart.reduce((max, item) => item.price > max.price ? item : max, cart[0]);
console.log(`Most expensive: ${mostExpensive.name} @ ₹${mostExpensive.price.toLocaleString()}`);

// Bonus: itemised bill
console.log("\n=== Cart Summary ===");
cart.forEach(item => {
  const subtotal = item.price * item.quantity;
  console.log(`${item.name.padEnd(10)} ₹${item.price} × ${item.quantity} = ₹${subtotal.toLocaleString()}`);
});
console.log("─".repeat(40));
console.log("TOTAL: ₹" + totalValue.toLocaleString());
````

## **Question 7:** Explain what array destructuring is and write code that extracts the first two colors from an array of colors into separate variables using destructuring.
// Array of colors

`const colors = ["red", "blue", "green", "yellow", "purple"]; `

**Answer:**

**Array destructuring** is an ES6 syntax that unpacks values from an array into distinct variables in a single statement, matching by **position**.

```js
const [first, second, ...rest] = array;
```

Benefits:
- Cleaner than `const first = arr[0]; const second = arr[1];`
- Can skip elements using commas: `const [, second] = arr;`
- Can capture remaining elements with the **rest pattern** (`...rest`)


````python
%%javascript
const colors = ["red", "blue", "green", "yellow", "purple"];

// Basic destructuring — extract first two
const [firstColor, secondColor] = colors;
console.log("First color: ", firstColor);   // red
console.log("Second color:", secondColor);  // blue

// Skip elements — get 1st and 3rd
const [primary, , third] = colors;
console.log("\n1st:", primary, " 3rd:", third);  // red  green

// Rest pattern — first two + remaining
const [c1, c2, ...remaining] = colors;
console.log("\nFirst two:", c1, c2);
console.log("Remaining: ", remaining);   // ["green", "yellow", "purple"]

// Swap two variables using destructuring (no temp variable needed!)
let a = "hello", b = "world";
[a, b] = [b, a];
console.log("\nSwapped: a =", a, " b =", b);
````

## **Question 8:** You have a student object with name, age, and marks properties. Write code using object destructuring to extract these values into separate variables and also rename the marks property to score while destructuring.
// Student object
```
const student = {
name: "Arjun",
age: 20,
marks: 85,
city: "Mumbai"
};
```

**Answer:**

Object destructuring unpacks properties by **key name** (not position). You can rename a property using `{ originalKey: newVariableName }`.



````python
%%javascript
const student = {
  name:  "Arjun",
  age:   20,
  marks: 85,
  city:  "Mumbai"
};

// Destructure and rename marks → score
const { name, age, marks: score, city } = student;

console.log("Name: ", name);   // Arjun
console.log("Age:  ", age);    // 20
console.log("Score:", score);  // 85  (renamed from marks)
console.log("City: ", city);   // Mumbai

// Bonus: default values during destructuring
const { grade = "Not assigned" } = student;
console.log("\nGrade:", grade);  // Not assigned (property doesn't exist)

// Bonus: nested object destructuring
const profile = { user: { name: "Riya", level: 3 } };
const { user: { name: userName, level } } = profile;
console.log("\nNested — User:", userName, " Level:", level);
````

## **Question 9:** Write a function called calculateTotal that can accept any number of prices as arguments and returns their sum. Use the rest operator to handle the variable number of parameters.
**Answer:**

The **rest operator** (`...params`) collects all remaining arguments into a **real array**, allowing a function to accept a variable number of parameters. It must always be the **last** parameter.

```js
function fn(first, ...rest) {
  // rest is an array of everything after the first argument
}
```

> Don't confuse rest (`...` in function params) with spread (`...` when calling a function or in array/object literals) — same syntax, different context.


````python
%%javascript
// Rest operator collects all arguments into the 'prices' array
function calculateTotal(...prices) {
  return prices.reduce((sum, price) => sum + price, 0);
}

// Works with any number of arguments
console.log(calculateTotal(100));                   // 100
console.log(calculateTotal(100, 250));              // 350
console.log(calculateTotal(100, 250, 500, 750));    // 1600
console.log(calculateTotal(100, 250, 500, 750, 1000)); // 2600

// Can also mix fixed + rest params
function applyDiscount(discount, ...prices) {
  return prices.map(p => p - p * (discount / 100));
}
console.log("\n10% off [200, 500, 1000]:", applyDiscount(10, 200, 500, 1000));
````

## **Question 10:** You have two arrays of student names and need to create a new array that contains all students from both arrays. Also, you have two objects representing a student's basic info and marks. Write code using the spread operator to solve both scenarios.
```
const classA = ["Rahul", "Priya", "Amit"];  
const classB = ["Sneha", "Vikram", "Pooja"];
```
**Answer:**

The **spread operator** expands an iterable (array/object) into individual elements. Used:
- In **array literals** `[...a, ...b]` → merge/copy arrays
- In **object literals** `{...obj1, ...obj2}` → merge/copy objects (later keys overwrite earlier ones)
- In **function calls** `fn(...arr)` → spread array as arguments


````python
%%javascript
const classA = ["Rahul", "Priya", "Amit"];
const classB = ["Sneha", "Vikram", "Pooja"];

// 1. Merge two arrays using spread
const allStudents = [...classA, ...classB];
console.log("All students:", allStudents);
// ["Rahul", "Priya", "Amit", "Sneha", "Vikram", "Pooja"]

// 2. Merge two objects using spread
const basicInfo = { name: "Arjun", age: 20, city: "Mumbai" };
const marks     = { maths: 90, science: 85, english: 78 };

const fullProfile = { ...basicInfo, ...marks };
console.log("\nFull profile:", fullProfile);

// 3. Copy an array without reference (spread creates a shallow copy)
const original = [1, 2, 3];
const copy     = [...original];
copy.push(99);
console.log("\nOriginal (unchanged):", original); // [1, 2, 3]
console.log("Copy (modified):      ", copy);      // [1, 2, 3, 99]

// 4. Spread to pass array as function arguments
const nums = [10, 5, 20, 3, 15];
console.log("\nMax in array:", Math.max(...nums)); // 20
console.log("Min in array:", Math.min(...nums)); // 3
````
