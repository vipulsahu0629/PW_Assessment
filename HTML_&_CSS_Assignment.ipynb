## **Question 1:** Create a layout with three boxes side by side using Flexbox. The first box should have twice the width of the other two. Provide the HTML and CSS code snippets.

**Answer:**

We use `display: flex` on the container and assign `flex: 2` to the first child and `flex: 1` to the other two.  
`flex: 2` means the first box takes **2 parts** of the total space, and the others each take **1 part** → ratio 2:1:1.

```css
.container { display: flex; gap: 10px; }
.box-wide  { flex: 2; }   /* twice as wide */
.box       { flex: 1; }   /* normal width  */
```


````python
%%html
<style>
  .flex-container {
    display: flex;
    gap: 12px;
    font-family: Arial, sans-serif;
    font-weight: bold;
    color: #fff;
    text-align: center;
  }
  .flex-container div {
    padding: 30px 10px;
    border-radius: 8px;
    font-size: 14px;
  }
  .box-wide { flex: 2; background: #2d6a4f; }
  .box-a    { flex: 1; background: #52b788; }
  .box-b    { flex: 1; background: #74c69d; }
</style>

<div class="flex-container">
  <div class="box-wide">Box 1<br>(flex: 2 — twice as wide)</div>
  <div class="box-a">Box 2<br>(flex: 1)</div>
  <div class="box-b">Box 3<br>(flex: 1)</div>
</div>
````

## **Question 2:** Using CSS Grid, create a 2×2 grid layout where the first row spans two columns, and the second row has two equal columns. Provide the HTML and CSS code snippets.

**Answer:**

We define a 2-column grid. The first item uses `grid-column: 1 / -1` (or `span 2`) to stretch across both columns. The remaining two items each occupy one column naturally.

```css
.grid-container {
  display: grid;
  grid-template-columns: 1fr 1fr;
  gap: 10px;

.header { grid-column: 1 / -1; }  /* spans both columns */
```


````python
%%html
<style>
  .grid-container {
    display: grid;
    grid-template-columns: 1fr 1fr;
    gap: 12px;
    font-family: Arial, sans-serif;
    font-weight: bold;
    color: #fff;
    text-align: center;
  }
  .grid-container div {
    padding: 28px 10px;
    border-radius: 8px;
    font-size: 14px;
  }
  .g-header { grid-column: 1 / -1; background: #1d3557; }
  .g-left   { background: #457b9d; }
  .g-right  { background: #a8dadc; color: #1d3557; }
</style>

<div class="grid-container">
  <div class="g-header">Row 1 — spans 2 columns (grid-column: 1 / -1)</div>
  <div class="g-left">Row 2 — Column 1</div>
  <div class="g-right">Row 2 — Column 2</div>
</div>
````

## **Question 3:** Write a media query that changes the background color of the body to light grey when the screen width is less than 600px.

**Answer:**

```css
/* Default body background */
body {
  background-color: white;
}

/* Applied only when viewport width < 600px */
@media (max-width: 599px) {
  body {
    background-color: lightgrey;
  }
}
```

**How it works:**
- `@media (max-width: 599px)` — the CSS rule inside fires only when the browser window (viewport) is 599 px wide or less.
- On wider screens the body keeps its default `white` background; on narrow/mobile screens it switches to `lightgrey`.
- Media queries are the core tool of **responsive design** — letting one HTML file look good on desktops, tablets, and phones.

````python
%%html
<style>
  .media-demo {
    padding: 20px;
    border-radius: 8px;
    font-family: Arial, sans-serif;
    background-color: white;
    border: 2px solid #ccc;
    text-align: center;
  }
  @media (max-width: 599px) {
    .media-demo { background-color: lightgrey; }
  }
</style>

<div class="media-demo">
  <p><strong>Resize your browser below 600px</strong> to see this box's background turn light grey!</p>
  <code>@media (max-width: 599px) { body { background-color: lightgrey; } }</code>
</div>
````

## **Question 4:** Create a button that changes its background color to blue and scales up slightly when hovered over, using CSS transitions. Provide the HTML and CSS code snippets.

**Answer:**

```css
.btn {
  background-color: #e63946;
  color: white;
  padding: 12px 28px;
  border: none;
  border-radius: 6px;
  cursor: pointer;
  transition: background-color 0.3s ease, transform 0.3s ease;
}

.btn:hover {
  background-color: #1d6ae5;   /* turns blue */
  transform: scale(1.1);       /* scales up 10% */
}
```

- `transition` lists **which properties** to animate and **how long** (0.3 s with an `ease` timing function).
- `:hover` defines the *end state*; the browser automatically interpolates between the normal and hovered states.

> **Hover over the button below to see it in action!**


````python
%%html
<style>
  .btn {
    background-color: #e63946;
    color: white;
    padding: 14px 32px;
    border: none;
    border-radius: 8px;
    cursor: pointer;
    font-size: 16px;
    font-family: Arial, sans-serif;
    font-weight: bold;
    transition: background-color 0.3s ease, transform 0.3s ease;
    display: inline-block;
    margin: 20px;
  }
  .btn:hover {
    background-color: #1d6ae5;
    transform: scale(1.1);
  }
</style>

<div style="text-align:center; padding:10px;">
  <button class="btn">Hover over me!</button>
</div>
````

## **Question 5:** Use the `::before` pseudo-element to add a decorative icon before each list item in an unordered list with the class name `decorated-list`. Provide the HTML and CSS code snippets.

**Answer:**

```css
.decorated-list {
  list-style: none;   /* remove default bullet */
  padding-left: 0;
}

.decorated-list li::before {
  content: "✅ ";    /* decorative icon injected before each item */
  color: green;
}
```

- `::before` creates a virtual element *before* the content of each `li`, with no extra HTML needed.
- `content` is the only required property for `::before` / `::after` — it defines what gets inserted (text, emoji, or even an image via `url()`).
- We remove the default bullet with `list-style: none` so the custom icon is the only visual marker.


````python
%%html
<style>
  .decorated-list {
    list-style: none;
    padding-left: 10px;
    font-family: Arial, sans-serif;
    font-size: 16px;
    line-height: 2;
  }
  .decorated-list li::before {
    content: "✅ ";
    color: green;
    font-size: 18px;
  }
</style>

<ul class="decorated-list">
  <li>Learn HTML basics</li>
  <li>Practice CSS Flexbox</li>
  <li>Understand CSS Grid</li>
  <li>Master Responsive Design</li>
</ul>
````

## **Question 6:** Create a simple animation that moves a square box from left to right infinitely. Provide the HTML and CSS code snippets.

**Answer:**

```css
/* 1. Define the keyframes */
@keyframes slideRight {
  from { transform: translateX(0); }
  to   { transform: translateX(300px); }
}

/* 2. Apply the animation to the element */
.moving-box {
  width: 60px;
  height: 60px;
  background-color: #e63946;
  animation: slideRight 2s linear infinite alternate;
}
```

Key `animation` shorthand values:
- `slideRight` — name of the `@keyframes` rule
- `2s` — duration of one cycle
- `linear` — constant speed (no ease-in/out)
- `infinite` — loops forever
- `alternate` — reverses direction every other cycle (left→right→left…), avoiding a sudden jump back


````python
%%html
<style>
  @keyframes slideRight {
    from { transform: translateX(0); }
    to   { transform: translateX(300px); }
  }

  .track {
    background: #f0f0f0;
    border-radius: 8px;
    padding: 20px 10px;
    width: 420px;
    overflow: hidden;
  }

  .moving-box {
    width: 60px;
    height: 60px;
    background: #e63946;
    border-radius: 6px;
    animation: slideRight 2s linear infinite alternate;
    display: flex;
    align-items: center;
    justify-content: center;
    color: #fff;
    font-size: 11px;
    font-family: Arial, sans-serif;
    font-weight: bold;
    text-align: center;
  }
</style>

<div class="track">
  <div class="moving-box">Box</div>
</div>
````

## **Question 7:** Explain the difference between `position: absolute;` and `position: fixed;` in CSS.

**Answer:**

| Feature | `position: absolute` | `position: fixed` |
|---------|----------------------|-------------------|
| **Positioned relative to** | Nearest *positioned* ancestor (an element with `position` ≠ `static`). Falls back to `<html>` if none exists. | The **browser viewport** (the visible window area). |
| **Scrolls with page?** | **Yes** — moves with the document when the user scrolls. | **No** — stays locked in place on screen regardless of scrolling. |
| **Removed from flow?** | Yes — other elements ignore it. | Yes — other elements ignore it. |
| **Common use cases** | Tooltips, badges, dropdown menus, labels placed inside a container. | Sticky navigation bars, floating action buttons, cookie banners, chat widgets. |

**Key mental model:**
- `absolute` → *"stick me to a corner of my parent box"*
- `fixed` → *"stick me to a corner of the screen"*


````python
%%html
<style>
  .pos-demo {
    position: relative;
    background: #d8f3dc;
    height: 120px;
    border: 2px solid #2d6a4f;
    border-radius: 8px;
    font-family: Arial, sans-serif;
    padding: 10px;
    margin-bottom: 16px;
  }
  .abs-badge {
    position: absolute;
    top: 8px; right: 8px;
    background: #e63946;
    color: #fff;
    padding: 4px 10px;
    border-radius: 20px;
    font-size: 12px;
    font-weight: bold;
  }
  .fixed-banner {
    position: fixed;
    bottom: 10px; right: 10px;
    background: #1d3557;
    color: #fff;
    padding: 8px 14px;
    border-radius: 8px;
    font-size: 12px;
  }
</style>

<div class="pos-demo">
  <p>Parent div (position: relative)</p>
  <p>The red badge is <strong>position: absolute</strong> — anchored to this parent's top-right corner.</p>
  <span class="abs-badge">Absolute Badge</span>
</div>

<div class="fixed-banner">📌 Fixed — stays on screen!</div>
````

## **Question 8:** Create a layered layout where a semi-transparent overlay covers an image using `z-index`. Provide the HTML and CSS code snippets.

**Answer:**

```css
.image-wrapper {
  position: relative;   /* creates the positioning context */
  width: 300px;
  height: 200px;
}
img {
  width: 100%; height: 100%;
  object-fit: cover;
  position: relative;
  z-index: 1;           /* image sits on layer 1 */
}
.overlay {
  position: absolute;   /* positions relative to .image-wrapper */
  top: 0; left: 0;
  width: 100%; height: 100%;
  background: rgba(0, 0, 0, 0.5);  /* semi-transparent black */
  z-index: 2;           /* overlay sits on layer 2 — above the image */
  display: flex;
  align-items: center;
  justify-content: center;
  color: white;
}
```

**How `z-index` works:**
- Elements with a higher `z-index` appear in front of elements with a lower value.
- `z-index` only works on elements that have a `position` value other than `static`.
- `rgba(0, 0, 0, 0.5)` means black with **50% opacity**, letting the image show through.


````python
%%html
<style>
  .image-wrapper {
    position: relative;
    width: 360px;
    height: 200px;
    border-radius: 10px;
    overflow: hidden;
    font-family: Arial, sans-serif;
    margin: 10px;
  }
  .image-wrapper img {
    width: 100%;
    height: 100%;
    object-fit: cover;
    position: relative;
    z-index: 1;
    display: block;
  }
  .overlay {
    position: absolute;
    top: 0; left: 0;
    width: 100%; height: 100%;
    background: rgba(0, 0, 0, 0.55);
    z-index: 2;
    display: flex;
    flex-direction: column;
    align-items: center;
    justify-content: center;
    color: #fff;
    text-align: center;
  }
  .overlay h2 { margin: 0 0 6px; font-size: 22px; }
  .overlay p  { margin: 0; font-size: 13px; opacity: 0.85; }
</style>

<div class="image-wrapper">
  <img src="https://picsum.photos/360/200?nature" alt="nature photo">
  <div class="overlay">
    <h2>Semi-Transparent Overlay</h2>
    <p>z-index: 2 &nbsp;|&nbsp; rgba(0,0,0,0.55)</p>
  </div>
</div>
````
