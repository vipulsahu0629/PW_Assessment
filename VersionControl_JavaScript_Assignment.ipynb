## **Question 1:** Explain the purpose of version control systems like Git. Why are they essential in modern software development?

**Answer:**

A **Version Control System (VCS)** is a tool that tracks every change made to files in a project over time, recording who changed what and when. **Git** is the most widely used distributed VCS today.

### What Git does:
- **Tracks history** — every saved state (commit) is stored permanently, so you can see the full evolution of your codebase and revert to any previous version if something breaks.
- **Branching** — developers can work on separate features or bug fixes in isolated branches without disturbing the main codebase, then merge them back when ready.
- **Collaboration** — multiple developers can work on the same project simultaneously. Git merges their changes and flags conflicts for manual resolution.
- **Distributed** — every developer has a full copy of the repository, so work continues even without internet access.

### Why Git is essential in modern development:

| Benefit | Why it matters |
|---------|---------------|
| **Undo mistakes** | Revert to any previous commit if a bug is introduced |
| **Parallel development** | Teams of hundreds work on the same codebase without overwriting each other |
| **Code review** | Pull Requests on GitHub let peers review changes before they reach production |
| **Deployment automation** | CI/CD pipelines trigger builds and tests automatically on every push |
| **Audit trail** | Every line of code is traceable to a commit, author, and reason (commit message) |

Without version control, teams would rely on emailing zip files or overwriting shared folders — error-prone, unscalable, and with no safety net.


## **Question 2:** List the typical steps you would follow when working with Git to make changes to a project and share them on GitHub.

**Answer:**

```bash
# Step 1 — Clone the repository (first time only)
git clone https://github.com/username/my-project.git
cd my-project

# Step 2 — Create a new branch for your feature / fix
git checkout -b feature-login

# Step 3 — Make your code changes (edit files in your editor)

# Step 4 — Check what has changed
git status

# Step 5 — Stage the files you want to include in the commit
git add index.html style.css          # specific files
# OR
git add .                             # all changed files

# Step 6 — Commit with a clear message
git commit -m "Add login form and basic validation"

# Step 7 — Push the branch to GitHub
git push origin feature-login

# Step 8 — Open a Pull Request on GitHub
# Go to github.com → your repo → "Compare & pull request"
# Team members review the code, leave comments, and approve.

# Step 9 — Merge into main after approval
git checkout main
git merge feature-login

# Step 10 — Pull latest changes before starting new work
git pull origin main
```

**Summary flow:**  
`clone` → `branch` → `edit` → `add` → `commit` → `push` → `Pull Request` → `merge`


## **Question 3:** What is the purpose of a `.gitignore` file in a Git repository? Provide an example of how to ignore all `.log` files.

**Answer:**

A **`.gitignore`** file tells Git which files and folders it should **not** track or include in commits. These are typically files that:
- Are automatically generated (build artifacts, compiled binaries)
- Contain sensitive data (API keys, passwords in `.env` files)
- Are machine-specific (IDE config files like `.vscode/`, `node_modules/`)
- Are large and unnecessary in the repo (log files, cache)

The `.gitignore` file lives in the root of the repository and uses simple pattern matching.

### Example — ignoring all `.log` files:

```
# .gitignore

# Ignore every file ending in .log (in any folder)
*.log

# Other common entries:
node_modules/        # npm dependencies
.env                 # environment variables / secrets
dist/                # build output folder
.DS_Store            # macOS system file
```

Once added to `.gitignore`, running `git status` will no longer show those files, and `git add .` will silently skip them.

> **Important:** If a file was already tracked before being added to `.gitignore`, you must untrack it with:  
> `git rm --cached filename.log`


## **Question 4:** You are currently on the `main` branch and want to switch to a branch named `feature-login`. Write the Git command to do this.

**Answer:**

```bash
# Option 1 — Modern syntax (Git 2.23+, recommended)
git switch feature-login

# Option 2 — Classic syntax (works in all Git versions)
git checkout feature-login
```

- Both commands move HEAD to `feature-login` and update your working directory to match that branch.
- If the branch doesn't exist yet locally but exists on GitHub, add `-t` (track):

```bash
git checkout -t origin/feature-login
# OR
git switch --track origin/feature-login
```

- To **create** the branch and switch to it in one step:

```bash
git checkout -b feature-login   # classic
git switch -c feature-login     # modern
```


## **Question 5:** Declare a variable named `age` using `let` and assign it the value `25`. Then, update its value to `30`.

**Answer:**

`let` declares a **block-scoped** variable that can be **reassigned** later (unlike `const`).


````python
%%javascript
// Declare and assign
let age = 25;
console.log("Initial age:", age);   // 25

// Reassign
age = 30;
console.log("Updated age:", age);   // 30
````

## **Question 6:** Declare a constant named `PI` with the value `3.14159`. Attempting to reassign it should result in an error. Demonstrate this behavior.

**Answer:**

`const` declares a **block-scoped constant** — it must be initialized at declaration and **cannot be reassigned**. Attempting to do so throws a `TypeError` at runtime.


````python
%%javascript
const PI = 3.14159;
console.log("PI =", PI);   // 3.14159

// Attempting to reassign a const — this will throw a TypeError
try {
  PI = 3;   // ❌ TypeError: Assignment to constant variable.
} catch (error) {
  console.error("Error caught:", error.message);
}

console.log("PI is still:", PI);  // 3.14159 — unchanged
````

## **Question 7:** You are given the following array:  
`const colors = ["red", "green", "blue", "yellow", "purple"];`

1. Access and print the **first** element.
2. Access and print the **last** element.
3. Access and print the **third** element.

**Answer:**

Arrays in JavaScript are **zero-indexed** — the first element is at index `0`.

| Element | Index | Expression |
|---------|-------|-----------|
| First | 0 | `colors[0]` |
| Last | length - 1 | `colors[colors.length - 1]` |
| Third | 2 | `colors[2]` |


````python
%%javascript
const colors = ["red", "green", "blue", "yellow", "purple"];

// 1. First element → index 0
console.log("First element:", colors[0]);               // red

// 2. Last element → index = length - 1
console.log("Last element:", colors[colors.length - 1]); // purple

// 3. Third element → index 2
console.log("Third element:", colors[2]);               // blue
````

## **Question 8:** Given `x = 10` and `y = 5`, write expressions to:
1. Add x and y
2. Check if x is greater than y
3. Check if x is equal to y

**Answer:**


````python
%%javascript
const x = 10;
const y = 5;

// 1. Addition operator (+)
const sum = x + y;
console.log("x + y =", sum);           // 15

// 2. Greater-than operator (>)  → returns a boolean
const isGreater = x > y;
console.log("x > y:", isGreater);      // true

// 3. Strict equality operator (===)  → checks value AND type
const isEqual = x === y;
console.log("x === y:", isEqual);      // false

// Extra: Why use === and not == ?
// == does type coercion:  "10" == 10  → true  (misleading!)
// === does not:           "10" === 10 → false (correct!)
console.log("\n'10' == 10  (loose):", "10" == 10);   // true  ⚠️
console.log("'10' === 10 (strict):", "10" === 10);    // false ✅
````

## **Question 9:** Identify the data types of the following values:
1. `"Hello"`
2. `42`
3. `true`
4. `undefined`

**Answer:**

JavaScript has 7 **primitive** data types. The `typeof` operator returns a string naming the type.

| Value | Data Type | Notes |
|-------|-----------|-------|
| `"Hello"` | **String** | Text enclosed in quotes (single, double, or backtick) |
| `42` | **Number** | JavaScript uses one `Number` type for all integers and decimals |
| `true` | **Boolean** | Only two values: `true` or `false` |
| `undefined` | **Undefined** | A variable declared but not yet assigned a value |

Other primitive types not shown here: `null`, `BigInt`, `Symbol`.


````python
%%javascript
// Using typeof to confirm data types at runtime
console.log(typeof "Hello");      // string
console.log(typeof 42);           // number
console.log(typeof true);         // boolean
console.log(typeof undefined);    // undefined

// Bonus examples
console.log("\n--- Additional examples ---");
console.log(typeof 3.14);         // number  (no separate float type)
console.log(typeof null);         // object  ⚠️ historic JavaScript bug — null is actually a primitive
console.log(typeof [1, 2, 3]);    // object  (arrays are objects)
console.log(typeof {name:"Raj"}); // object
console.log(typeof function(){}); // function
````
